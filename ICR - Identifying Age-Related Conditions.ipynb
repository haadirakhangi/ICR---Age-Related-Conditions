# MODEL TRAINING ON THE ORIGINAL DATASET

### Training the model without scaling and hyperparameter tuning

```python
x = df_tr.drop('Class',axis=1)
y = df_tr['Class']
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import make_pipeline

models = [LogisticRegression(max_iter=5000),RandomForestClassifier(n_estimators=100), KNeighborsClassifier(n_neighbors=10), DecisionTreeClassifier(criterion = 'gini'), SVC(kernel='rbf')]

score= []

for model in models:
    cvs = cross_val_score(model, x, y, cv = StratifiedKFold(n_splits=5)).mean()
    score.append(
    {
        'model_name': model,
        'score': cvs
    })
```

```python
score = pd.DataFrame(score)
score
```

### Training the model on scaled data

```python
def get_scores(model, X_train, X_test, y_train, y_test):
    pipe = make_pipeline(RobustScaler(), model)
    pipe.fit(X_train, y_train)
    return pipe.score(X_test, y_test)
```

```python
skf = StratifiedKFold(n_splits=5)

pipe_score = []

pipe_score_mean = []

for model in models:
    for train_index, test_index in skf.split(x,y):
        X_train, X_test = np.array(x)[train_index], np.array(x)[test_index]
        y_train, y_test = np.array(y)[train_index], np.array(y)[test_index]
        pipe_score.append(get_scores(model, X_train, X_test, y_train, y_test))
    pipe_score_mean.append(
    {    
        'model_name': model,
         'score': np.array(pipe_score).mean() 
    })
```

```python
pipe_score = pd.DataFrame(pipe_score_mean)
pipe_score
```

### Scaling the data and using hyperparameter tuning to train the model

```python
model_params = {
    'Logistic':{
        'model': LogisticRegression(max_iter=10000),
        'params': {
            'C': range(1,20)
        }
    },
    'SVM':{
        'model': SVC(gamma='scale', kernel='rbf'),
        'params': {
            'C': range(1,20),
        }
    },
    'RFC':{
        'model': RandomForestClassifier(criterion='gini'),
        'params': {
            'n_estimators': [100,200,400,500],
        }
    },
    'KNN':{
        'model': KNeighborsClassifier(metric='minkowski'),
        'params': {
            'n_neighbors': range(3,20)
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV

final_score = []
for model_name, mp in model_params.items():
    x_robust = RobustScaler().fit_transform(x)
    rscv = GridSearchCV(mp['model'],mp['params'], cv= StratifiedKFold(n_splits=5))
    rscv.fit(x_robust,y)
    final_score.append({
        'model_name': model_name,
        'best_parameters': rscv.best_params_,
        'best_score': rscv.best_score_
    })
```

```python
final_score = pd.DataFrame(final_score)
final_score
```
